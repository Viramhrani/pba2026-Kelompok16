# Sentiment Analysis dengan PyCaret

Tujuan:
- Membandingkan beberapa algoritma machine learning
- Memilih model terbaik untuk klasifikasi sentimen
- Menyimpan model untuk deployment

## 1. Import Library

```python
import pandas as pd
from pycaret.classification import *
```

## 2. Load Dataset Hasil Preprocessing

```python
data = pd.read_csv('../data/ml_reviews_ready.csv')
data.head()
```

## 3. Cek Missing Value

```python
data.isnull().sum()
# Kalau ada nilai kosong:
data = data.dropna()
```

## 4. Setup PyCaret

```python
clf = setup(
    data=data,
    target='sentiment',
    text_features=['clean_text'],
    session_id=42,
    train_size=0.8,
    fold=5,
    verbose=False
)
```

## 5. Benchmark 3 Model

```python
lr = create_model('lr')
nb = create_model('nb')
rf = create_model('rf')
```

## 6. Bandingkan Semua Model

```python
best_model = compare_models(include=['lr', 'nb', 'rf'])
```

## 7.Evaluasi Model Terbaik

```python
evaluate_model(best_model)
```

## 8. Prediksi pada Data Baru

```python
new_data = pd.DataFrame({
    'clean_text': [
        'game nya seru banget dan skin nya keren',
        'server sering lag dan banyak bug setelah update'
    ]
})

hasil_prediksi = predict_model(best_model, data=new_data)
hasil_prediksi
```

## 9. Simpan Model Terbaik

```python
save_model(best_model, '../app/best_model')
```
